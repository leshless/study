### Задание 1
Изобразить на рисунках область определения функций

1) $z=\ln(x^2+y)$

2) $z=\sqrt{\sin(x^2+y^2)}$

3) $z=\sqrt{x^2-4}+\sqrt{9-y^2}$

```python
x, y = sp.symbols("x y")

# 1
D = (x**2 + y >= 0)
sp.plot_implicit(D)

# 2
D = (sp.sin(x**2 + y**2) >= 0)
sp.plot_implicit(D)

# 3
Dx = (x**2 - 4 >= 0)
Dy = (9 - y**2 >= 0)
sp.plot_implicit(sp.And(Dx, Dy))
```

### Задача 2.
Построить на отдельных графиках линии уровня функций $z=\ln(x^2+y)$ и $z=x^2-y^2$.

```python
f = lambda x, y : sp.log(x**2 + y)

p = sp.plot(show=False)
for c in [-2, -1, 0, 1, 2]:
  p.append(sp.plot_implicit(sp.Eq(f(x, y), c), show=False)[0])
p.show()

f = lambda x, y : x**2 - y**2

p = sp.plot(show=False)
for c in [-2, -1, 0, 1, 2]:
  p.append(sp.plot_implicit(sp.Eq(f(x, y), c), show=False)[0])
p.show()
```

### Задача 4.
Найти точки разрыва функции $z=\frac1{(x - y^2)^2}$, изобразить их на графике.

```python
x, y = sp.symbols("x y")

eq = sp.Eq((x - y**2)**2, 0)
display(sp.solve(eq))

p = sp.plot(show=False)
p.append(sp.plot_implicit(eq, show=False)[0])

p.show()
```

### Задача 5.
Найти частные производные по каждому аргументу функции $u=(xy)^z$

```python
x, y, z = sp.symbols("x y z")

u = (x * y) ** z

for s in (x, y, z):
  display(u.diff(s))
```

### Задача 6.
Проверить выполнение теоремы Эйлера для функции $f(x,y)=\frac{x}{x^2+y^2} $.
###### Указание.
Упростить выражение с помощью simplify, прежде чем сравнивать.

```python
x, y = sp.symbols("x y")
k = sp.symbols("k")

f = x / (x**2 + y**2)
display(f.subs({x : k*x, y : k*y}).factor(k))
# dim = -1

assert (x * f.diff(x) + y * f.diff(y)).simplify() == -1 * f
```

### Задача 7.
Найти полный дифференциал функции $z=\sin^2x+\cos^2y$

```python
x, y = sp.symbols("x y")
dx, dy = sp.symbols("dx dy")

z = sp.sin(x)**2 + sp.cos(y)**2
display(z.diff(x)*dx + z.diff(y)*dy)
```

```python
f = lambda x, y : -sp.asin(4*x*y)

x, y = sp.symbols("x y")

p = sp.plot(show=False)
for c in [-1, -0.5, 0, 0.5, 1]:
  p.append(sp.plot_implicit(sp.Eq(f(x, y), c), show=False)[0])
p.show()

dx, dy = sp.symbols("dx dy")
display(f(x, y).diff(x)*dx + f(x, y).diff(y)*dy)
```
