### Задание 1.
Даны комплексные числа вида
$$
z_k = k + \frac{i}{k},\quad k = 1, 2, 3.
$$
Вычислить (представить в алгебраической форме) с помощью sympy
$$
z_k \cdot z_{k + 1},\quad z_k^2,\quad \sqrt{z_k}.
$$

```python
k = sympy.Symbol("k")
zk = k + I / k

for kval in [1, 2, 3]:
  display(zk.subs(k, kval))
  display(zk.subs(k, kval) * zk.subs(k, kval + 1))
  display(zk.subs(k, kval) ** 2)
  display(sympy.sqrt(zk.subs(k, kval)))
```

### Задание  3.
Изобразить на координатной плоскости точки, соответствующие корням  $3$-й степени комплексного числа $z = -\frac{1}{2} + i\frac{\sqrt{3}}{2}$. Подписать точки $A_k$, где $k$  - номер корня.

```python
z = complex(-0.5, np.sqrt(3)/2)

roots = [z**(1/3) * np.exp(2j * np.pi * k / 3) for k in range(3)]

plt.figure(figsize=(6, 6))
for i in range(3):
    root = roots[i]
    plt.plot(root.real, root.imag, 'o', label=f'$A_{k+1}$')
    plt.text(root.real, root.imag, f'$A_{k+1}$', fontsize=12, ha='right')

plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.xlabel('Re')
plt.ylabel('Im')

plt.legend()
plt.show()
```

### Задание  4.
Изобразить на координатной плоскости точки, соответствующие $z_4 = - \sqrt{3} + i$ и $\bar{z_4}$, подписать $z_4$ тригонометрической формой, $\bar{z_4}$ - экспоненциальной ($z = r e^{i\varphi}$, $r$ и $\varphi$ такие же, как в тригонометрической форме).

```python
z4 = complex(-np.sqrt(3), 1)
z4_ = complex(-np.sqrt(3), -1)

r = abs(z4)
phi = np.angle(z4)

# Построение на координатной плоскости
plt.figure(figsize=(6, 6))
plt.plot(z4.real, z4.imag, 'o', label=f'$z_4 = {r:.0f}(cos({phi:.2f}) + i sin({phi:.2f}))$')
plt.plot(z4_.real, z4_.imag, 'o', label=f'$\overline{{z_4}} = {r:.2f}e^(-i{phi:.2f})$')

plt.text(z4.real, z4.imag, f'$z_4$', fontsize=12, ha='right')
plt.text(z4_.real, z4_.imag, f'$\overline{{z_4}}$', fontsize=12, ha='right')

# Оформление графика
plt.xlim(-4, 1)
plt.ylim(-2, 2)
plt.xlabel('Re')
plt.ylabel('Im')
plt.legend()
plt.show()
```

#### Задание  5*.
Изобразить корни 4 степени числа $5e^{\frac{i\pi}{8}}$, в полярной системе координат, подписи к точкам - аргументы (углы) сответствующих комплексных чисел.

Число вводить в тригонометрической форме, для получения корней комплексного числа $z = r(\cos(\varphi) + \sin(\varphi))$ использовать формулу
$\sqrt[n]{z} = \sqrt[n]{r}(\cos(\varphi_k) + i \sin(\varphi_k))$,
$\varphi_k = \frac{\varphi + 2\pi k}{n}$, $k = 0, ..., n$.

Вывести алгебраическую форму полученных корней на экран и сравнить с результатом, полученным с помощью sympy.root (с использованием sympy.expand_complex для упрощения выражения). Аналитическое представление оказывается разным.

Для каждого корня найти разность в формате float модуля разности корней, полученных двумя способами (т.е. нужно для каждой пары одинаковых корней, полученных двумя способами находить разность корней, вычислять модуль sympy.Abs этой разности и преобразовывать результат в формат float).

```python
r = 5
phi = np.pi / 8
n = 4

def roots(r, phi, n):
    return [r**(1/n) * (np.cos((phi + 2 * np.pi * k) / n) + 1j * np.sin((phi + 2 * np.pi * k) / n)) for k in range(n)]

analytical_roots = roots(r, phi, n)

z_sym = sympy.re(r)*sympy.exp(sympy.I*phi)
sympy_roots = [sympy.expand_complex(sympy.root(z_sym, n, k)) for k in range(n)]

plt.figure(figsize=(8, 8))
for k in range(len(analytical_roots)):
    root = analytical_roots[k]
    angle = np.angle(root)
    plt.polar([0, angle], [0, abs(root)], marker='o', label=f'arg: {angle:.2f}')

plt.legend()
plt.show()

for k in range(n):
    analytical_real = analytical_roots[k]
    sympy_real = complex(sympy_roots[k].evalf())

    print(f'{k+1} аналитически: {analytical_real}')
    print(f'{k+1} sympy: {sympy_real}')

    diff_mod = abs(abs(analytical_real) - abs(sympy_real))
    print(f'diff: {diff_mod:.20f}\n')
```

```python
import sympy

z = sympy.symbols('z')
equation = z**5 + z**4 * (-7 + 8*sympy.I) + z**3 * (-30 - 46*sympy.I) + z**2 * (254 - 44*sympy.I) + z * (-484 + 420*sympy.I) + 640 - 440*sympy.I

roots = sympy.solve(equation, z)
display(roots)

modulos = [abs(root.evalf()) for root in roots]
display(modulos)

prod = sympy.N(sympy.prod(roots))
display(abs(prod))
display(sympy.arg(product))
```
